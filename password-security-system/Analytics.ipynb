```python
# Import 
import pandas as pd
import matplotlib.pyplot as plt

print("Import Successful")
```

```python
df = pd.read_csv("validation_history.csv")

df["timestamp"] = pd.to_datetime(df["timestamp"])
df
```

```python
# Pie Chart (Strong vs Weak)
df["strength_group"] = df["strength_label"].apply(
    lambda x: "Strong" if x in ["Strong", "Very Strong"] else "Weak"
)

counts = df["strength_group"].value_counts()

plt.figure()
plt.pie(
    counts.values,
    labels=counts.index,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Strong vs Weak Passwords Tested")
plt.show()
```

```python
# Strength Distribution (Histogram)
plt.figure()
plt.hist(df["strength_score"], bins=10)
plt.xlabel("Strength Score")
plt.ylabel("Frequency")
plt.title("Password Strength Distribution")
plt.show()
```

```python
# Line chart (Strength Over Time)
plt.figure()
plt.plot(df["timestamp"], df["strength_score"], marker="o")
plt.xlabel("Time")
plt.ylabel("Strength Score")
plt.title("Password Strength Improvement Over Time")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
